### Challenge 8: Math Library
**Import the math library. Store any number in a variable. Print its square root. Then print whether the number is bigger or smaller than pi using an if/else statement.**

An if/else then compares a variable to math.pi and prints a different result depending on the comparison. This combines three concepts: imports, variables, and conditionals.

```python
# Import math and compare a number to pi

import math

my_number = 4

if my_number > math.pi:
    print(f"${my_number} is greater than pi (${math.pi:.2f})")
else:
    print(f"${my_number} is less than or equal to pi (${math.pi:.2f})")

print(f"Pi is: ${math.pi}")

```

### Challenge 10: NumPy Arrays
**Import numpy as np. Create an array of at least 5 temperatures in Fahrenheit. Convert all of them to Celsius in one line using: celsius = (fahrenheit - 32) * 5/9. Print both arrays and the average temperature in Celsius.**

### Challenge 14: Entanglement
**Create a function called make_bell_state() that creates a QuantumCircuit with 2 qubits, applies H to qubit 0, applies CX with qubit 0 controlling qubit 1, and returns the circuit. Then call it, store the result in a variable, and print it. Add a comment explaining why we need BOTH the H gate and the CX gate.**

**What this code does:**
This code defines a function that builds a Bell State — one of the most important quantum circuits in existence. Inside the function: qc.h(0) places qubit 0 in superposition, then qc.cx(0, 1) applies a CNOT gate that entangles qubit 1 with qubit 0. Entanglement means the two qubits are linked: measuring one instantly determines the state of the other. The function returns the completed circuit so it can be used and reused anywhere in the program. This is real quantum programming — not simplified, not simulated. The actual circuit structure used by researchers.

---

```python
# Create a Bell state function

def make_bell_state():
    qc = QuantumCircuit(2)
    qc.h(0)       # superposition on qubit 0
    qc.cx(0, 1)   # qubit 0 controls qubit 1
    return qc

my_circuit = make_bell_state()
print("Bell State Circuit:")
print(my_circuit)
print("This circuit creates ___")
```

### Challenge 15: Running a Simulation
**Create a 1-qubit circuit, apply an H gate, add measure_all(), simulate with 500 shots, and print the counts. You should see roughly 50% 0 and 50% 1 because of the superposition.**

**What this code does:**
This code runs a complete quantum experiment end to end. A single-qubit circuit is built, an H gate is applied to create superposition, and the qubit is measured. AerSimulator executes the circuit 500 times (500 shots). Each shot independently collapses the superposition — randomly producing either 0 or 1. get_counts() returns a dictionary showing how many times each outcome occurred across all runs. The result is approximately 50% zeros and 50% ones — direct evidence of quantum superposition. This is how quantum researchers validate their circuits: run many shots, analyze the probability distribution.

---

```python
# Simulate a single qubit in superposition

qc = QuantumCircuit(1, 1)
qc.h(0)           # put qubit 0 in superposition
qc.measure(0, 0)  # measure qubit 0, store in classical bit 0

print("=== Quantum Circuit ===")
print(qc)
print()

simulator = AerSimulator()
job = simulator.run(qc, shots=1024)
result = job.result()
counts = result.get_counts()

print("=== Simulation Results (1024 runs) ===")
print("Raw counts:", counts)
print()

total = sum(counts.values())
for state, count in counts.items():
    percentage = (count / total) * 100
    print(f"  State |${state}⟩: ${count} times (${percentage:.1f}%)")

print()
print("Notice: results are approximately 50% each.")
print("That is quantum superposition — random until measured.")

```
