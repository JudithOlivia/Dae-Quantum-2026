## The H-Z-H Pattern

**Step 1: First H gate does...**

It puts the qubit into superposition

**Step 2: Z gate does...**

It performs a phase flip, which adds a mathematical "negative" to the wave if the qubit is in the 1 state.

**Step 3: Second H gate does...**

It acts as the interference creator, bringing the superposed waves back together to collide.

**The result is:**

The quibit is forced to be in a definite state. so 0 or 1.

# PART 2: DEUTSCH-JOZSA ALGORITHM

## The Problem It Solves

**What is the mystery box problem?**
Imagine you are given a "black box" (an oracle) that takes an input (like 0 or 1) and gives you a result. You are told the box is either Constant or Balanced, and you have to figure out which one it is. A classical computer might have to check the box multiple times, but a quantum computer uses interference to figure it out in just one try.

**Constant function means:**

The box is "stuck" and gives the exact same output no matter what you feed into it. Whether you give it all 0s or all 1s, it will always return only 0s or only 1s.

**Balanced function means:**

The box is "fair" and gives a 0 for exactly half of the possible inputs and a 1 for the other half.

---

## Classical vs Quantum Solution

### Classical Computer Approach:

**How many checks does a classical computer need?**

more than half

**Why can't it do better?**

A classical computer has to check inputs one by one, and if the first few results are all the same, you still don't know if the next one will be different.

---

### Quantum Computer Approach:

**How many checks does a quantum computer need?**

one

**How is this possible?**

By using superposition, the computer tests all possible inputs at the same time, creating a "wave" of all potential answers.
